```python
list_certified=[]
for i in visa_df['continent'].unique():
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Certified' # For certified employees
    con=con1 & con2
    list_certified.append(len(visa_df[con]))
    #print(f"The no of employees of {i} people 'certified visa' are {len(visa_df[con])} ")
list_certified
```

```python
list_denied=[]
for i in visa_df['continent'].unique():
    con1=visa_df['continent']==i
    con2=visa_df['case_status']=='Denied'  # For denied employees
    con=con1 & con2
    list_denied.append(len(visa_df[con]))
    #print(f"The no of employees of {i} people with 'Denied visa' are {len(visa_df[con])} ")
list_denied
```

```python
d={'Certified':list_certified, 'Denied':list_denied}
data=pd.DataFrame(d,index=visa_df['continent'].unique())
data
```

- It will give the amount of relation between variables
- It is denoted by r
- -1<r<1
- For +ve relation : r--> 0 to 1
- For -ve relation : r--> -1 to 0
- For no relation : r --> 0
- In python we have, **corr** function under pandas
    - This will give covariance matrix
    - In this data(visa_df) we have 3 numerical values
- So we get 9 values
