**Data Transformation**

- Generally data has 3 types of distribution
    - Normal distribution
    - Positive skewed
    -  Negative skewed
- All the maths concept developed by assumption as Data follows Normal distribution
- Data Skew happens because of many reasons, in that one reason is outliers
- Once we perform the outlier analysis then we can apply transformation methods, if the data does not follows normality
- We have some transformation methods like
    - Log transformation
    - Exponential transformation
    - Reciprocal transformation
    - Square root transformation
    - Power transformation

$Step-1$

**Import packages**

```python

import numpy as np
import matplotlib.pyplot as plt
```

$Step-2$

**Read the data**

```python
exp_data = np.random.exponential(size=10000)
exp_data
```

$Step-3$

**Plot the data**

```python
plt.hist(exp_data,bins = 20,label ='Exponential') # bins = intervals
```

```python
norm_data = np.random.normal(size=10000)
plt.hist(norm_data,bins=20,label='Normal')
plt.show()
```

**Goal**
  
**Convert exp_data to Normal distribution**

$Step-4$

**Log Transformation**

- It means performing logarithm operation on original data
- It is an approach to convert into normal distribution
- Log means natural logarithm (base=e)
- loge(x) = 2.3 * log10(x)

```python
log_data = np.log(exp_data)
plt.hist(log_data,bins=20,label='Logarithms')
plt.show()
```

```python
plt.figure(figsize=(14,5))
plt.subplot(1,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(log_data,bins=20,label='Logarithms')
plt.legend() # for showing label
plt.show()
```

**Reciprocal transformation**

- Reciprocal will fail when values are low
- Eg: x=0 then 1/0 fails

```python
x=2
1/x
```

```python
reci_data = 1/exp_data
plt.figure(figsize=(14,5))
plt.subplot(1,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(1,2,2).hist(reci_data,bins=20,label='Reciprocal')
plt.legend() # for showing label
plt.show()
```

**Square root Transformation**

```python
exp_data = np.random.exponential(size=10000)
sq_rt_val = np.sqrt(exp_data)
plt.figure(figsize=(15,5))
plt.subplot(2,2,1).hist(exp_data,bins=20,label='Exponential')
plt.legend()
plt.subplot(2,2,2).hist(sq_rt_val,bins=20,label='Square root')
plt.legend() # for showing label
plt.subplot(2,2,3).hist(log_data, bins=20,label='Logarithms')
plt.legend()
plt.subplot(2,2,4).hist(reci_data, bins=20, label='Reciprocal')
plt.legend()
plt.show()
```

**Power Transformation**

- Box-Cox
- Ye-johnson
- It is under sklearn package

```python
from sklearn.preprocessing import PowerTransformer
#save the package
power=PowerTransformer(method='yeo-johnson')
# apply fit transform
exp_data_1 = exp_data.reshape(-1,1)
power.fit_transform(exp_data_1)
# power transformer expecting every values in one list
```

```python
from sklearn.preprocessing import PowerTransformer
exp_data = np.random.exponential(size=10000)
exp_data_1 = exp_data.reshape(-1,1)
plt.figure(figsize=(14,5))

#Box-cox
power = PowerTransformer(method='box-cox')
box_cox = power.fit_transform(exp_data_1)
plt.subplot(1,2,1).hist(box_cox,bins=20,label='Box-Cox')
plt.legend()

#Yeo-johnson
power=PowerTransformer(method='yeo-johnson')
yeo_johnson = power.fit_transform(exp_data_1)
plt.subplot(1,2,2).hist(yeo_johnson,bins=20,label='Yeo-Johnson')
plt.legend()
plt.show()
```
