- Bcz in ML models the input is in form of numbers
- ML models developed by maths, So passing text data will not works
- We have different types of methods are there to convert categorical to numerical
    - Map method
    - np.where
    - One hot encoder
    - Label encoder

**Task**

- Read the data
- Apply for loop, iterate through columns
- Get unique labels of each labels
- Get count of unique labels
- Create a dict
- Apply into map

```python
visa_df=pd.read_csv(r"C:\Users\Lenovo\Python NareshIT\Data files\Visadataset - Visadataset.csv")
for i in list(visa_df.select_dtypes(include='object').columns):
    labels = list(visa_df[i].unique())
    values = [i for i in range(len(labels))]
    d = dict(zip(labels,values))
    visa_df[i] = visa_df[i].map(d)
visa_df
```

**One hot Encoder**

- One hot means if one is ON another one is OFF
- On --> 1 and OFF --> 0
- Eg: 'case_status' has 2 unique labels('Denied', 'Certified')
- If apply this on case_status it will create 2 extra columns like case_status_denied & case_status_certified

| case_status|case_status_denied|case_status_certified|
|---------|--|--|
| Denied|1|0|
| Certified|0|1|

**Advantages**

- In ML we have one imp property is **Independent of variables**
- ML models expect that variables are **independent each other**
- One hot encoder creating new columns, which are **independent each other**
- Independent variables means both var have **90 degree phase shift**
- 90 degree phase shift --> perpendicular each other --> orthogonal each other 

**Diasdvantages**

- One hot encoder creates new columns = no of unique labels
- If no of unique labels increased then new columns also increased
- The more columns lead more data
- More data --> complexity increased --> more computation power require --> time increased
- The entire process named as **Curse of dimensionality**
